```python
# in this code an example of fitting with numpy and pytorch will be developed
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
x_data = np.arange(1, 6)
y_data = x_data * 2 + 0.9
print(x_data)
```

```python
plt.scatter(x_data, y_data)
```

```python
def forward(x, w, b=0):
    y_hat = x * w + b
    return y_hat
```

```python
def loss(y_hat, y):
    return (y_hat - y) ** 2
```

```python
all_w = []
all_loss = []

for w in np.arange(0, 5, 0.1):
    l_sum = 0
    
    for i in range(len(x_data)):
        y_hat = forward(x_data[i], w, 1)
        l = loss(y_hat, y_data[i])
        
        l_sum += l
    
    all_w.append(w)
    all_loss.append(l_sum/len(y_data))
```

```python
plt.stem(all_w, all_loss)
```
